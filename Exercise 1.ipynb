```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from datetime import datetime
```

```python
with open('sx-stackoverflow-a2q.txt') as f:
    a2q = f.readlines()
```

```python
with open('sx-stackoverflow-c2q.txt') as f:
    c2q = f.readlines()
```

```python
with open('sx-stackoverflow-c2a.txt') as f:
    c2a = f.readlines()
```

```python
G = nx.DiGraph()
```

```python
len(a2q)
```

```python
t_start = datetime(2008, 10, 1)
t_end = datetime(2008, 11, 1)

for line in a2q[:10000] :
    line = line.replace('\n', '')
    line = line.split(' ') 
    t = datetime.fromtimestamp(int(line[2]))
    if t_start <= t < t_end :
        if G.has_edge(line[0], line[1]) :
            G[line[0]][line[1]]['weight'] += 1
            G[line[0]][line[1]]['timestamp'].append(datetime.fromtimestamp(int(line[2])))
        else :
            G.add_edge(line[0], line[1], weight=1)
            G[line[0]][line[1]]['timestamp']=[datetime.fromtimestamp(int(line[2]))]

for line in c2q[:10000] :
    line = line.replace('\n', '')
    line = line.split(' ') 
    t = datetime.fromtimestamp(int(line[2]))
    if t_start <= t < t_end :
        if G.has_edge(line[0], line[1]) :
            G[line[0]][line[1]]['weight'] += 1
            G[line[0]][line[1]]['timestamp'].append(t)
        else :
            G.add_edge(line[0], line[1], weight=1)
            G[line[0]][line[1]]['timestamp']=[t]

for line in c2a[:10000] :
    line = line.replace('\n', '')
    line = line.split(' ') 
    t = datetime.fromtimestamp(int(line[2]))
    if t_start <= t < t_end :
        if G.has_edge(line[0], line[1]) :
            G[line[0]][line[1]]['weight'] += 1
            G[line[0]][line[1]]['timestamp'].append(datetime.fromtimestamp(int(line[2])))
        else :
            G.add_edge(line[0], line[1], weight=1)
            G[line[0]][line[1]]['timestamp']=[datetime.fromtimestamp(int(line[2]))]
        
```
